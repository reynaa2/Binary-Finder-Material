```python
import matplotlib.pyplot as plt
import numpy as np
import apogee.tools.read as apread
import pylab as pl
from BFData import *
from BinaryFinderTool import *
import os
```

```python
def R_set(locationID,apogeeID,ranger,value):
    R_name = []
    R_Peak = []
    j = []
    for i in range(visit):
        #print(locationID[i],apogeeID[i])
        if locationID[i] != '1':
            folder = '/Volumes/CoveyData/APOGEE_Spectra/APOGEE2_DR13/Bisector/BinaryFinder2/'
            filename = '/Volumes/CoveyData/APOGEE_Spectra/APOGEE2_DR13/Bisector/BinaryFinder2/{0}/{1}/{2}.tbl'.format(ranger, locationID[i], apogeeID[i])
            if os.path.exists(filename):
                name = BFData(locationID[i],apogeeID[i],ranger)
                R_name.append(name.lowestR(value))
                R_Peak.append(name.lowestRPeak())
            else:
                j.append([locationID[i],apogeeID[i]])
    return R_name, R_Peak
```

```python
plotname = 'Ranger 0.1_'+ str(locationIDs) +'_'+ str(apogeeIDs)
example = BFData('4102', '2M21394731+4249041', 0.1)
TargetCount = 1
apogeeIDs=example.apogeeID
locationIDs=example.locationID
print(locationIDs)
print(example.r)
print(example.r[0][0])
print(example.r[1][1])

fig = plt.figure(figsize=(10,10))

for i in range(TargetCount):
    badheader, header = apread.apStar(locationIDs,apogeeIDs, ext=0, dr='13', header=True)
    z= apread.apStar(locationIDs, apogeeIDs, ext=9, header=False, dr='13')
    nvisits = header['NVISITS']
    for visit in range(0, nvisits):
        if (nvisits != 1):
            CcF = z['CCF'][0][2+ visit]
        else:
            CcF = z['CCF'][0]       
        #Conditions to adjust the legend size. For R_1 - R_5 Window Calculations      

        sub1 = fig.add_subplot(221)
        sub1.plot(CcF + visit,label= 'Visit: '+str(1+visit))
        sub1.axvspan(10,391,color= 'green', alpha=0.05)
        sub1.set_title('All Visits for '+ str(locationIDs)+'/'+str(apogeeIDs),fontsize=12)
        sub1.set_ylabel('$\widehat{CCF}$ Units', fontsize=12)
        sub1.set_xlabel('CCF Lag',fontsize=12)
        sub1.text(10,2.7,'Min R of all visits for $R_1 =$ '+str(round(min(R1),3)), fontsize=12)
        sub2 = fig.add_subplot(222)
        sub2.plot(CcF + visit,label= 'Visit: '+str(1+visit))
        sub2.axvspan(20,381,color= 'green', alpha=0.05)
        sub2.text(10,2.7,'Min R of all visits for $R_2 = $'+str(round(min(R2),3)),fontsize=12)
        sub2.set_ylabel('$\widehat{CCF}$ Units', fontsize=12)
        sub2.set_xlabel('CCF Lag',fontsize=12)
        sub3 = fig.add_subplot(223)
        sub3.plot(CcF + visit,label= 'Visit: '+str(1+visit))
        sub3.axvspan(30,371,color='green', alpha=0.05)
        sub3.text(10,2.7,'Min R of all visits for $R_3 = $'+str(round(min(R3),3)),fontsize=12) 
        sub3.set_ylabel('$\widehat{CCF}$ Units', fontsize=12)
        sub3.set_xlabel('CCF Lag',fontsize=12)
        sub4 = fig.add_subplot(224)
        sub4.plot(CcF + visit,label= 'Visit: '+str(1+visit))
        sub4.axvspan(40,361,color='green', alpha=0.05)
        sub4.text(10,2.7,'Min R of all visits for $R_4 = $'+str(round(min(R4),3)),fontsize=12)
        #sub4.set_ylabel('$\widehat{CCF}$ Units', fontsize=12)
        sub4.set_xlabel('CCF Lag',fontsize=12)
        sub1.legend(loc = 'lower right',fontsize=10)
        plt.savefig('/Volumes/CoveyData/APOGEE_Spectra/APOGEE2_DR13/Bisector/BinaryFinder_Plots/HD_0.1/'+str(plotname)+'.png')    
plt.show()
```
